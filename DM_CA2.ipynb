Pre-processing is one of the most important steps in data mining projects. First we load csv file as a DataFrame using pandas library.

```python
pd.reset_option('display.max_rows')
df = pd.read_csv("Market_Basket.csv", header = None)
df
```

```python
L = df.values.tolist()
dataset = [[i for i in x if pd.notnull(i)] for x in L]
dataset
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
te_ary
```

```python
te.columns_
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
sum_dataset = df.sum()
plt.figure(figsize=(200,100))
ax = sns.barplot(sum_dataset.index, sum_dataset , alpha=0.8, color='r')
ax.set_title("Performance in Inhibiting Corona Virus",fontsize=25)
ax.set_xlabel('Location',fontsize=25)
ax.set_ylabel('Performance',fontsize=25)
ax.set_xticklabels(sum_dataset.index, rotation='vertical', fontsize=25)
plt.xticks(rotation = 80)
plt.show()
```

## Question 3

### A

```python
def find_frequent_itemsets(_min_support, min_lenght):
    frequent_itemsets = apriori(df, min_support=_min_support, use_colnames=True)
    frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
    return frequent_itemsets[ (frequent_itemsets['length'] == min_lenght)]
```

## Question 4

```python
def find_association_rules(_min_support, _min_confidence):
#     frequent_itemsets = apriori(df, min_support=_min_support, use_colnames=True)
    frequent_itemsets = fpgrowth(df, min_support = _min_support, use_colnames=True)
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=_min_confidence)
    return rules.sort_values(by=['lift'], ascending=False)   
```
